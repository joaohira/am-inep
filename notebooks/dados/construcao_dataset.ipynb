# Construção do dataset

```python
import os
import pandas as pd
import numpy as np
```

```python
DATA_PATH = '../../dados/brutos/'
```

```python
# Carregar os dados do ENEM por Escola, comum para todos os censos
EPE_PATH = os.path.join(DATA_PATH, 'enem_por_escola.csv')
epe = pd.read_csv(EPE_PATH, sep=';', na_values='.', encoding='windows-1252', low_memory=False)
```

```python
# Carregar os dados do Censo de 2010
CENSO_2015_PATH = os.path.join(DATA_PATH, 'censo_2015.csv')
censo = pd.read_csv(CENSO_2015_PATH, sep='|', na_values='.', encoding='windows-1252', low_memory=False)
```

```python
# Selecionar os dados de 2015
epe_2015 = epe[epe['NU_ANO'] == 2015]
```

```python
# Realizar o inner join entre as duas tabelas
merged_2015 = epe_2015.merge(censo.rename(columns={'CO_ENTIDADE':'CO_ESCOLA_EDUCACENSO'}), how='inner', on='CO_ESCOLA_EDUCACENSO')
```

```python
# Exportar o .csv finalizado
merged_2015.to_csv(os.path.join(DATA_PATH, 'merged_2015.csv'), index=False)
```
